```python
from bs4 import BeautifulSoup 
import urllib
```

```python
link_int = "https://www.bankbazaar.com/reviews.html?reviewPageNumber="
```

```python
url = link_int + str(1)
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}
req = urllib.request.Request(url = url, headers = headers)
```

```python
doc = urllib.request.urlopen(req).read().decode("utf-8")
```

```python
doc
```

```python
bs0bj= BeautifulSoup(doc)
```

```python
only_reviews = bs0bj.find_all(name = "div" , attrs ={'class':'text_here review-desc-more'})
```

```python
only_reviews_txts = [x.text for x in only_reviews]
only_reviews_txts
```

```python
containers = bs0bj.find_all(name= 'li', attrs={'class':'review-box'})
containers_0 = containers[0]
```

```python
containers_0
```

```python
xx = containers_0.find(name='a', attrs ={'class':'user-review-comment js-individual-title'})
xx
```

```python
title = xx.get_attribute_list('title')[0].strip()
```

```python
title
```

```python
rating =  containers_0.find(name='input').get_attribute_list('value')[0]
rating
```

```python
profile = containers_0.find(name='div',attrs ={'class':'reviewer-profile'}).text

profile
```

```python
import re
```

```python
up =re.sub('Reviewed on','', profile)
up
```

```python
up= re.sub('\\n','',up)
up
```

```python
name= re.findall("^[A-Z ]+",up)[0].strip()
name
```

```python
date = re.findall("[A-Za-z]{3} [0-9]{2}, [0-9]{4}",up)[0]
date
```

```python
up1= re.sub("Mar 21, 2020", "", up)
up1
```

```python
place = re.findall("(?<=, )[A-Za-z ]+", up1)[0].strip()
place
```
